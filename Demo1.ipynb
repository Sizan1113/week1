```python
#demo
marks = int(input("Enter your marks:"))
if (marks > 100):
    print("Invalid")
elif (marks >=90 and marks<=100):
    print("A+")
elif (marks >= 80 and marks <90):
    print("A")
elif (marks >= 70 and marks <80):
    print("B+")
elif (marks >= 60 and marks <70 ):
    print("B")
elif (marks >= 50 and marks <60 ):
    print("C+")
elif (marks >= 40 and marks <50 ):
    print("C")
elif ( marks <40 ):
    print("NG")


else:
    print("fail")
    
```

```python
#demo 1
#input marks of 5 subject
#calculate total and average
#print grade as well on the basis of average
maths_marks = float(input("Enter your marks for Maths:"))
science_marks = float(input("Enter your marks for Science:"))
computer_marks = float(input("Enter your marks for Computer:"))
english_marks = float(input("Enter your marks for English:"))
nepali_marks = float(input("Enter your marks for Nepali:"))
total = maths_marks + science_marks + computer_marks + english_marks + nepali_marks
print(f"\n total marks:",total)
avg_marks = (total)/5
print(f"\n average marks:",avg_marks)
if (avg_marks > 100):
    print("Invalid")
elif (avg_marks >=90 and avg_marks<=100):
    print("A+")
elif (avg_marks >= 80 and avg_marks <=89.9):
    print("A")
elif (avg_marks >= 70 and avg_marks <=79.9):
    print("B+")
elif (avg_marks >= 60 and avg_marks <=69.9 ):
    print("B")
elif (avg_marks >= 50 and avg_marks <=59.9 ):
    print("C+")
elif (avg_marks >= 40 and avg_marks <=49.9 ):
    print("C")
elif (avg_marks <40 ):
    print("NG")
else:
    print("fail")
    
    
```
