Indexing

```python
num = [1,2,3,4,56,7]
print(num[1:3])
```

```python
#slicing
num = [1,2,3,4,5,6]
print(num[-3:-1])
print(num[0:5:2])
print(num[:5])
print(num[0:])
print(num[::2])
print(num[-5:-1:2])
```
