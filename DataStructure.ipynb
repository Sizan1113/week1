```python
#we cannot define empty set
#for example
n = {}
type(n)
```

```python
#for creating an empty set we need to use se() function
empty_set = set()
type(empty_set)
```

```python
#Set removes duplicate data and arrange in ascending order
num = [1,2,10,20,20,11,23]
num_set = set(num)
print(num_set)
```
