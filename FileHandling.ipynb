Text file read and write

```python
file = open("data.txt","w")#IT accepts two arguments -> file name and operstion to perform
file.write("Hello World!")
file.close()
```

```python
file = open("data.txt","r")
content = file.read()
print(content)
file.close()
```

```python
file = open("data.txt","a")
file.write("I am Sizan")
file.close()
```

#Read and Write CSV File.

```python
import csv
countries = [
    ["country","Destination"],
    ["Nepal","Mustang"],
    ["India","Delhi"],
    ["France","Paris"]
]
with open("country.csv","w",newline = "")as file:
    writer = csv.writer(file)
    writer.writerows(countries)
```

```python
with open("country.csv","r") as file:
    reader = csv.reader(file)
    for row in reader:
        
        print(row)
```

Dictionary Read and Write

```python
personal_details = [
    {
        "name":"Ram",
        "age":20
    },{
        "name":"Hari",
        "age":19
    },{
        "name":"Shyam",
        "age":18
    }
]
with open("personal_details.csv","w", newline = "") as file:
    columns = ["name","age"]
    writer = csv.DictWriter(file,fieldnames = columns)
    writer.writeheader()
    writer.writerows(personal_details)
```

```python
with open("personal_details.csv","a", newline = "") as file:
    columns = ["name","age"]
    writer = csv.DictWriter(file,fieldnames = columns)
    writer.writeheader()
    writer.writerows(personal_details)
```

```python
with open("personal_details.csv","r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
```

```python
#Read and write json data
import json
info = {
    "name":"Ram",
    "age":30
}
with open("info.json","w") as file:
    json.dump(info,file,indent = 10)
```

```python
with open("info.json","r") as file:
    data = json.load(file)
print(data)
```
