#Function

```python
#to define a function we use a key word "def"
#There are two type of functions in python one with parameter and arguments and vice versa
#An example of non-parameterised function
def greet():
    return"hello world"
greet()
```

```python
#parameterised function
def welcome(name):
    print(f"welcome {name}")
welcome("Sizan")
```

```python
def add(a,b):
    return a + b
add(5,10)
```

#Default argument and Keyword argument

```python

def person(name,age,address):
    print(f"I am {name}.I am {age}years old and i live at {address}")
person("Ram",20,"Baneswor")
```

```python
#Default argument
def person(name = "Sizan",age="18",address="Nepal"):
    print(f"I am {name}.I am {age}years old and i live at {address}")
person()
```

```python
def validator(email="please enter your email ",password="please enter your password"):
    print(f"your email:{email}.\n your password:{password}")
validator()
```

```python
def employees(name,department,salary,bonus):
    print(f"Name of employee is:{name}\nDepartment:{department}\nsalary:{salary}\nbonus:{bonus}")
employees("Ram","Sales",20000,"10%")
```

```python
#keyword argument
employees("Sales","Ram","sizan",1000)
```

```python
employees(department = "sales",name="Ram",bonus = "10%",salary = 30000)
```

Lambda

```python
#syntax
#function = lambda arguments : expression
add = lambda a,b : a + b
print(add(1,2))
```

*args

```python
def car(car_name,*args):
    print("Name of car is",car_name)
    
    for detail in args:
        print(detail)
    
```

```python
car("Hyundai","Creta",1992)
```

Keyword arguments

```python
def car(car_name,**details):
    print("Name of car is",car_name)
    
    for key,value in details.items():
        print(f"{key}:{value}")
car("Honda", model = "city",buildyear = 1992)  
```

```python
def student_details(**details):
    for key,value in details.items():
        print(f"{key}:{value}")
student_details(name="Ram",gender = "M", address ="baneswor",phone =989899898)
```

#Recursion


```python
#countdown
def countdown(n):
    if n == 6:
        return
    else:
        print(n)
        countdown(n+1)
countdown(1)       
```

#Nested function


```python
def calculate(a,b):
    def add(a,b):
        return a+b
    def substract(a,b):
        return a-b
    def multiply(a,b):
        return a + b
    def divide(a,b):
        return a/b
    return add(a,b), substract(a,b), multiply(a,b), divide(a,b)
result = calculate(1,2)
print("sum is:",result[0])
print("diffrence is:",result[1])
print("product is:",result[2])
print("divide  is:",result[3])
```
