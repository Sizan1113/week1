#For loop

```python
#Syntax
#for (variable) in range(start,stop(n-1),step):
#print(variable)
for a in range (1,11,2):
    print(a)
```

```python
a = "Sizan"
for i in a:
    print(i)
```

```python
#Syntax for iterating over a list
#for variable in list:
#print(variable)

list_of_foods = ["momo","pizza","burger"]
for i in list_of_foods:
    print(i)
```

```python
list_of_foods = ["momo","pizza","burger"]
for i in list_of_foods:
    print(i)
    for j in i:
        print (j)
```

```python
#iterating over a tuple
num = (1,2,3,4,5)
for a in num:
    print (a)
```

```python
for i in range(1,101):
    if i % 2 == 0:
        print(i)
```

```python
sum = 0
for i in range(1,11):
    sum = sum + i
    print(sum)
```

```python
sum = 0
for i in range(1,11):
    sum += i
print(f"The total sum is:",sum)
```

```python
#print multiples of a num usinf for loop
num = int(input("Enter a number:"))
product = 1
for num in range(1,11):
    for i in range(1,11):
        product = num * i
        print(f"{num}*{i}={product}" ) 
```

```python
#print multiples of a num usinf for loop
num = int(input("Enter a number:"))
product = 1
for i in range(1,11):
    product = num * i
    print(f"{num}*{i}={product}" ) 
```
