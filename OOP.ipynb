OOP(Object oriented programming)

```python
'''
class => Its an blueprint of an object
Object =>functions methods'''
'''
class contains attributes and methods
class are accessed by creating an object'''
'''
for eg:
class       Object
fruit       apple,banana
car          name,model'''
```

```python
class Dog:
    species = "Mammal"
    def sound(self):
        return "Bark"
d1 = Dog()
d1.sound()
        
```

```python
d1.species
```

```python
d2 = Dog()
print(d2.sound())
d2.species
```

```python
class Person:
    Gender1 = "Male"
    Gender2 = "Female"
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
p1 = Person("Ram",24)
print(p1.name)
print(p1.age)
print(p1.Gender1)
```

```python
p2=Person("Sita",20)
print(p2.name)
print(p2.age)
print(p2.Gender2)
```

```python
class Laptop:
    ram = "16GB"
    cpu = "i7 14gen"
    size = "14inch"
    def __init__(self,name,model,screen,keyboard):
        self.name = name
        self.model = model
        
        self.screen = screen
        self.keyboard = keyboard
    def start(self):
        return "Laptop Started"
    def stop(self):
        return "Laptop shutdown"
l1 = Laptop("Dell","Inspiron","LED","Half")
print(f"You have a {l1.name} laptop of {l1.model} model which has a {l1.screen} screen and {l1.keyboard} keyboard with specs like {l1.ram} RAM ,{l1.cpu}processor and {l1.size} size")
print(l1.start())
print(l1.stop())
```
