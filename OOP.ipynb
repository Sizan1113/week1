```python
class Laptop:
    ram = "16GB"
    cpu = "i7 14gen"
    size = "14inch"
    def __init__(self,name,model,screen,keyboard):
        self.name = name
        self.model = model
        
        self.screen = screen
        self.keyboard = keyboard
    def start(self):
        return "Laptop Started"
    def stop(self):
        return "Laptop shutdown"
l1 = Laptop("Dell","Inspiron","LED","Half")
print(f"You have a {l1.name} laptop of {l1.model} model which has a {l1.screen} screen and {l1.keyboard} keyboard with specs like {l1.ram} RAM ,{l1.cpu}processor and {l1.size} size")
print(l1.start())
print(l1.stop())
```

```python
'''
4 pillars of OOP
Inheritance : object from child class can access methods from parent class
Encapsulation : data hiding -> for security .
Polymorphism :Method override
Abstraction:data hiding -> for data logic implementation.
'''
```

Single Inheritance

```python
class Vehicle:
    def start(self):
        return "Vehicle Stopped!!"
    def go(self):
        return "Vehicle moving!!"
    def stop(self):
        return "Vehicle stopped!!"
class Bike(Vehicle):
    def accelerate(self):
        return"Go!"
b1 = Bike()
b1.accelerate()
        
```

```python
b1.go()
```

Multiple Inheritance

```python
class A:
    def function_A(self):
        return "value A"
class B:
    def function_B(self):
        return"value B"
class C:
    def function_C(self):
        return "value C"
class D(A,B,C):
    pass
d = D()
print(d.function_A())
print(d.function_B())
print(d.function_C())
```

Multilevel Inheritance

```python
class Vehicle:
    def start(self):
        return "Vehicle Started!!"
    def stop(self):
        return"Vehicle Stopped!!"
class Two_Wheeler(Vehicle):
    def accelerate(self):
        return "Go forward!!"
    def brake(self):
        return "Stop!!"
class Bike(Two_Wheeler):
    def __init__(self,name,model,buildyear):
        self.name = name
        self.model = model
        self.buildyear = buildyear
b1 = Bike("BMW","1000rr",1990)
print(b1.name)
print(b1.model)
print(b1.buildyear)
print(b1.start())
print(b1.brake())
        
```

#Encapsulation


```python
#Data hiding
#Data Security
#access modifiers ->private (__)
                    #protected(_)
                    #and public
```

```python
#Declaring and accessing private variable.
class Person:
    def __init__(self,name,gender):
        self.__name = name
        self.__gender = gender
p = Person("Ram","M")
print(p._Person__name)
print(p._Person__gender)
```

```python
class Person:
    def __init__(self,name,gender):
        self.__name = name
        self.__gender = gender
    def print_details(self):
        print(f"Name of the person:{self.__name}")
        print(f"Gender :{self.__gender}")
p = Person("Ram","M")
print(p.print_details())
```

```python
class Personal_Details:
    def __init__(self,name,gender):
        self._name = name
        self._gender = gender
    def personal_details_function(self):
        print(f"Name of customer:{self._name}")
        print(f"Gender of Customer:{self._gender}")
pd = Personal_Details("Ram","Male")
pd.personal_details_function()
```

#Polymorphism

```python
#Duck Typing
class Circle:
    def area(self,r):
        return 3.14 * r*r
class Rectangle:
    def area(self,l,b):
        return l * b

C1 = Circle()
R1 = Rectangle()
print(C1.area(7))
R1.area(5,4)
    
```

```python
#Duck Typing
class Circle:
    def __init__(self,r):
        self.r = r
    def area(self):
        return 3.14 * self.r*self.r
class Rectangle:
    def __init__(self,l,b):
        self.l =l
        self.b = b
    def area(self):
        return self.l * self.b
def findarea(shape):
    print(shape.area())
    

C1 = Circle(4)
R1 = Rectangle(3,5)
findarea(R1)
findarea(C1)    
```
